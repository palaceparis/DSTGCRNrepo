```python
import requests
import pandas as pd
```

```python
us_raw = pd.read_csv(
    "/home/rbmg6-mingze/Projects/DSTGCRN/data/raw/carbonmonitor-us_datas_2023-07-04.csv"
)
```

```python
us_raw
```

```python
us_raw["date"] = pd.to_datetime(us_raw["date"], format="%d/%m/%Y")  # year format
us_raw["value"] = pd.to_numeric(us_raw["value"])  # number format

print(us_raw.dtypes)
```

```python
us_raw = (
    us_raw.groupby(["state", "date"])["value"].sum().reset_index(name="mt_co2_per_day")
)  # emissions from the sum of all sectors
us_raw
```

```python
data_mat = us_raw.pivot(
    index="state", columns="date", values="mt_co2_per_day"
).reset_index()
data_mat.columns.name = None
data_mat = data_mat.rename_axis(None, axis=1)
data_mat.set_index("state", inplace=True)  # Set states as index
data_mat
```

```python
# Exclude "United States"
data_mat_filtered = data_mat[~data_mat.index.isin(["United States"])]
data_mat_filtered_transposed = data_mat_filtered.transpose()
data_mat_filtered_transposed
```

```python
data_mat_filtered_transposed.to_csv(
    "/home/rbmg6-mingze/Projects/DSTGCRN/data/interim/us_emi.csv", index=True
)
```
