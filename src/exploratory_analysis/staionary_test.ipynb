```python
from statsmodels.tsa.stattools import adfuller
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "/home/rbmg6-mingze/Projects/Emissions-Prediction/data/interim/emissions.csv",
    index_col=0,
)
```

```python
# Apply lgo to df
df_log = np.log(df)
df_log
```

```python
# ADF test
for province in df_log.columns:
    result = adfuller(df_log[province])
    print(f"Province: {province}")
    print("ADF Statistic: %f" % result[0])
    print("p-value: %f" % result[1])
    print("-----------------------------")
```

```python
# ADF test
for province in df.columns:
    result = adfuller(df[province])
    print(f"Province: {province}")
    print("ADF Statistic: %f" % result[0])
    print("p-value: %f" % result[1])
    print("-----------------------------")
```

```python
non_stationary_provinces = ["Beijing", "Jilin", "Heilongjiang", "Tibet"]

for province in non_stationary_provinces:
    df[province] = df[province].diff()

# # drop the first row which is NaN
df = df.dropna()
```

```python
# ADF test again
for province in df.columns:
    result = adfuller(df[province])
    print(f"Province: {province}")
    print("ADF Statistic: %f" % result[0])
    print("p-value: %f" % result[1])
    print("-----------------------------")
```
