```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Model": ["Emissions", "Emissions and Trend", "Full Model"],
    "MAE": [0.019, 0.012, 0.011],
    "MAPE": [0.02328, 0.01565, 0.0141],
    "RMSE": [0.036, 0.022, 0.02],
    "RMSPE": [0.03915, 0.02512, 0.02169],
    "R_squared": [0.931, 0.973, 0.978],
}

labels = data["Model"]
metrics = list(data.keys())[1:-1]  # Get the metric names (excluding R_squared)
r_squared = list(data.keys())[-1]  # R_squared is plotted on a separate axis
bar_width = 0.15
opacity = 0.8

# Setting the positions of bars on X axis
r = np.arange(len(labels))

fig, ax1 = plt.subplots()

for i, metric in enumerate(metrics):
    values = data[metric]
    ax1.bar(r + i * bar_width, values, width=bar_width, alpha=opacity, label=metric)

# Adding xticks
ax1.set_ylabel("Metrics", fontweight="bold")
ax1.set_xticks(
    r + 2 * bar_width
)  # Position of the xticks in the middle of the group bars
ax1.set_xticklabels(labels)

# Add R_squared on a second y-axis
ax2 = ax1.twinx()
ax2.plot(
    r, data[r_squared], marker="o", color="purple", label="$R^2$"
)  # R_squared displayed as "RÂ²"
ax2.set_ylabel("$R^2$", fontweight="bold")

# Combine legends
fig.legend(loc="upper right", bbox_to_anchor=(1, 0.93), bbox_transform=ax1.transAxes)
plt.savefig(
    "/home/rbmg6-mingze/Projects/Emissions-Prediction/data/processed/comparison_feature.png",
    dpi=300,
)
plt.show()
```
