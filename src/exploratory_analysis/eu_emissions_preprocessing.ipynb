```python
import requests
import pandas as pd
```

```python
eu_raw = pd.read_csv(
    "/home/rbmg6-mingze/Projects/DSTGCRN/data/raw/carbonmonitor-eu_datas_2023-07-04.csv"
)
```

```python
eu_raw
```

```python
eu_raw["date"] = pd.to_datetime(eu_raw["date"], format="%d/%m/%Y")  # year format
eu_raw["value"] = pd.to_numeric(eu_raw["value"])  # number format
print(eu_raw.dtypes)
```

```python
eu_raw = (
    eu_raw.groupby(["country", "date"])["value"]
    .sum()
    .reset_index(name="mt_co2_per_day")
)  # emissions from the sum of all sectors
eu_raw
```

```python
data_mat = eu_raw.pivot(
    index="country", columns="date", values="mt_co2_per_day"
).reset_index()
data_mat.columns.name = None
data_mat = data_mat.rename_axis(None, axis=1)
data_mat.set_index("country", inplace=True)  # Set states as index
data_mat
```

```python
# Exclude "EU27 & UK"
data_mat_filtered = data_mat[~data_mat.index.isin(["EU27 & UK"])]
data_mat_filtered_transposed = data_mat_filtered.transpose()
data_mat_filtered_transposed
```

```python
data_mat_filtered_transposed.to_csv(
    "/home/rbmg6-mingze/Projects/DSTGCRN/data/interim/eu_emi.csv", index=True
)
```
